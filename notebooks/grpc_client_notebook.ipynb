```python
import grpc
import ml_service_pb2
import ml_service_pb2_grpc

def run_grpc_examples():
    channel = grpc.insecure_channel('localhost:50051')
    stub = ml_service_pb2_grpc.MLServiceStub(channel)
    
    # Health check
    health_response = stub.HealthCheck(ml_service_pb2.HealthRequest())
    print(f"Health: {health_response.status}")
    
    # Get model classes
    classes_response = stub.GetModelClasses(ml_service_pb2.ModelClassesRequest())
    print(f"Available classes: {classes_response.classes}")
    
    # List datasets
    datasets_response = stub.ListDatasets(ml_service_pb2.DatasetsRequest())
    print("Datasets:")
    for dataset in datasets_response.datasets:
        print(f"  - {dataset.name} ({dataset.size} bytes)")

if __name__ == '__main__':
    run_grpc_examples()
```
